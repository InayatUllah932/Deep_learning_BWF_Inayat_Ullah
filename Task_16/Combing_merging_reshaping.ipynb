### Merging 

```python
pd.merge(df1,df2)
```

#### In the above code I do not specify which column to combine

```python
pd.merge(df1,df2,on="key")
```

#### While in the above code I specify to combin on key column

```python
df3 = pd.DataFrame({'lkey': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],'data1': range(7)})
```

```python
df4 = pd.DataFrame({'rkey': ['a', 'b', 'd'],'data2': range(3)})
```

```python
df3
```

```python
df4
```

#### If the column names are different in each object, you can specify them separately:

```python
pd.merge(df3,df4,left_on="lkey",right_on="rkey")
```

#### In the above code I merge the two file on lkey and rkey columns. This type of join called inner join. In inner join Only the common items fron the two files are selected

```python
pd.merge(df1, df2, how='outer')
```

#### In the above code the two files merge each other.Which is called Full outer merge

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

####  Merging on Index

```python
left1 = pd.DataFrame({'key': ['a', 'b', 'a', 'a', 'b', 'c'],'value': range(6)})
```

```python
left1
```

```python
right1 = pd.DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

#### The left_on parameter specifies the column name in the left DataFrame to use for the merge, which is "key" in this case. The right_index parameter specifies to use the indices of the right DataFrame for the merge instead of a specific column.This is Inner join

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
left = pd.DataFrame({'key1': ['Ohio', 'Ohio', 'Ohio','Nevada', 'Nevada'],
                      'key2': [2000, 2001, 2002, 2001, 2002],
                      'data': np.arange(5.)})
```

```python
left
```

```python
right = pd.DataFrame(np.arange(12).reshape((6, 2)),
    index=[['Nevada', 'Nevada', 'Ohio', 'Ohio','Ohio', 'Ohio'],
    [2001, 2000, 2000, 2000, 2001, 2002]],
    columns=['event1', 'event2'])
```

```python
right
```

```python
pd.merge(left,right,left_on=["key1","key2"],right_index=True)
```

```python
pd.merge(left,right,left_on=["key1","key2"],right_index=True,how="outer")
```

#### Using the indexes of both sides of the merge is also possible:


```python
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
    index=['a', 'c', 'e'],
    columns=['Ohio', 'Nevada'])
```

```python
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
    index=['b', 'c', 'd', 'e'],
    columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2,right2,left_index=True,right_index=True)
```

```python
pd.merge(left2,right2,left_index=True,right_index=True,how="outer")
```

#### You can join ot merge the two datafram through the following code

```python
left2.join(right2, how='outer')
```

### Reshaping 

```python
data = pd.DataFrame(np.arange(6).reshape((2, 3)),
 index=pd.Index(['Ohio', 'Colorado'], name='state'),
 columns=pd.Index(['one', 'two', 'three'],
 name='number'))
```

```python
data
```

```python
result = data.stack()
```

```python
result
```

#### stack() method in pandas DataFrame is used to transform a DataFrame from a "wide" format to a "long" format by pivoting the columns into rows.

```python
result.unstack()
```

#### The above code unstack the stack code

```python
s1 = pd.Series([0, 1, 2, 3], index=['a', 'b', 'c', 'd'])
```

```python
s2 = pd.Series([4, 5, 6], index=['c', 'd', 'e'])
```

```python
s1
```

```python
s2
```

```python
data2 = pd.concat([s1, s2], keys=['one', 'two'])
```

```python
data2
```

```python
data2.unstack()
```

```python
data2.unstack()
```

```python
data2.unstack().stack(dropna=False)
```

#### You can drop the null data by dropna = True

```python
data2.unstack().stack(dropna=True)
```
