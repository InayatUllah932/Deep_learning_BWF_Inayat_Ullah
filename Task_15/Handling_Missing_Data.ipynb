## another data frame

```python
data = pd.DataFrame([[1., 6.5, 3.], [1., np.nan, np.nan],[np.nan, np.nan, np.nan], [np.nan, 6.5, 3.]])
```

```python
data
```

```python
data.dropna(how='all')
```

#                                 Removing Duplicates

```python
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],'k2': [1, 1, 2, 3, 3, 4, 4]})
```

```python
data
```

## Drop the duplicates values in the file 

```python
data.drop_duplicates()
```

## To add new column

```python
data["k3"] = range(7)
```

```python
data
```

## To drop duplicate data in particular columns

```python
data['k1'].drop_duplicates()
```

## To keep the last column as it is

```python
data.drop_duplicates(['k1', 'k2'], keep='last')
```

## Replacing 1 with 6 let suppose

```python
data.replace(1,6)
```

## You can replace list of items with list of items

```python
data.replace([1,6],[0.5,0.7])
```

# Outlier

```python
data = pd.DataFrame(np.random.randn(1000, 4))
```

```python
col = data[2]
```

```python
col
```

```python
col[np.abs(col) > 3]
```

```python
data[(np.abs(data) > 3).any(1)]
```
